```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Example with FLAN-T5 Large
model_name = "google/flan-t5-large"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
model.save_pretrained(r"C:\Users\srich\OneDrive\Desktop\new\assignment")
tokenizer.save_pretrained(r"C:\Users\srich\OneDrive\Desktop\new\assignment")
```

```python
from llmsherpa.readers import LayoutPDFReader
import os

def process_pdf(pdf_path):

    llmsherpa_api_url = "http://localhost:5001/api/parseDocument?renderFormat=all"
    pdf_reader = LayoutPDFReader(llmsherpa_api_url)
    

    abs_path = os.path.abspath(pdf_path)
    print("Absolute Path:", abs_path)

    
    with open(pdf_path, 'rb') as pdf_file:
        pdf_content = pdf_file.read()  
    pdf_url=f"file://{abs_path}"
    doc = pdf_reader.read_pdf( pdf_url,contents=pdf_content)
    return doc

def query_llm(content, query):
    
    input_text = f"Context: {content} \n\nQuestion: {query}"
    print(input_text)
    print(content)

    inputs = tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)
    

    outputs = model.generate(
        inputs["input_ids"],
        max_length=1024,
        num_beams=4,
        early_stopping=True
    )
    
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
def split_content(content, chunk_size=512):
    
    words = content.split()
    for i in range(0, len(words), chunk_size):
        yield " ".join(words[i:i+chunk_size])


if __name__ == "__main__":
    pdf_file_path = "Iterate Interplay- Sample Order Form .pdf"
    query = "give a summary of the order form"
    
    try:
        
        doc = process_pdf(pdf_file_path)
        
        
        chunks = doc.chunks()
        content = "\n".join([chunk.to_context_text() for chunk in chunks])
        print("Extracted Content Length:", len(content))
        
        
        chunk_size = 512  
        responses = []
        for chunk in split_content(content, chunk_size):
            print("Processing Chunk (Preview):", chunk[:100])  #-- used for debugging
            response = query_llm(chunk, query)
            responses.append(response)
        
        
        final_response = " ".join(responses)
        print("Final Response:", final_response)
        
    except FileNotFoundError as e:
        print(f"File Not Found: {e}")
    except AttributeError as e:
        print(f"Attribute Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

```
